```python
import pandas as pd

# ===== CONFIGURACIÓN =====
archivo = "blastp_results.csv"
min_alignment = 500
min_identity = 25
max_evalue = 1e-5

# ===== CARGAR ARCHIVO =====
df = pd.read_csv(archivo, header=None)

# Ahora sí 13 columnas
df.columns = [
    "query",
    "accession",
    "identity",
    "alignment_length",
    "mismatches",
    "gap_opens",
    "q_start",
    "q_end",
    "s_start",
    "s_end",
    "evalue",
    "bit_score",
    "positives"
]

print(f"Total filas originales: {len(df)}")

# ===== ELIMINAR DUPLICADOS POR ACCESSION =====
df_unique = df.drop_duplicates(subset="accession")
print(f"Accesiones únicas: {len(df_unique)}")

# ===== APLICAR FILTROS =====
df_filtered = df_unique[
    (df_unique["alignment_length"] >= min_alignment) &
    (df_unique["identity"] >= min_identity) &
    (df_unique["evalue"] <= max_evalue)
]

print(f"Accesiones que cumplen filtros: {len(df_filtered)}")

# ===== GUARDAR RESULTADO =====
df_filtered[["accession"]].to_csv("TPS_filtradas.txt", index=False, header=False)

print("Archivo 'TPS_filtradas.txt' generado.")
```
