```python
import re
from collections import defaultdict
from Bio import SeqIO

archivo = "TPS_all_sequences.fasta"

# Expresión regular para DDXXD
patron = re.compile(r"D{2}[A-Z]{2}D")

motivos_encontrados = defaultdict(int)
total_motivos = 0

for record in SeqIO.parse(archivo, "fasta"):
    secuencia = str(record.seq)
    
    matches = patron.findall(secuencia)
    
    for motivo in matches:
        motivos_encontrados[motivo] += 1
        total_motivos += 1

# Resultados
print("\nMotivos DDXXD encontrados:\n")

for motivo, count in sorted(motivos_encontrados.items()):
    print(f"{motivo} → {count} veces")

print("\nResumen:")
print(f"Motivos distintos: {len(motivos_encontrados)}")
print(f"Total de ocurrencias: {total_motivos}")
```
